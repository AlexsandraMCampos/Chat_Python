```python
import pandas as pd
import numpy as np
import matplotlib as plt

# Criando DataFrame (matriz com os dados das semanas)
dados = np.array([[160, 120,90, 110, 110], [70, 150, 90, 140, 70],[90, 100, 80, 80, 70], [100, 150, 90, 130, 70]])

df =pd.DataFrame(dados, index = ["1ª Semana", "2ª Semana","3ª Semana","4ª Semana"] , columns= ["2ªf","3ªf","4ªf","5ªf","6ªf"])


#  Cálculo do desvio padrão das visitas por semana
# axis=1 operação será realizada nas colunas, que neste array são cada linha (semana)
#.round(2) função que arredonda o resultado para 2 casas decimais
desvio_padrao = df.std(axis=1).round(2)

# Cálculo da média e mediana por semana
media = df.mean(axis=1).round(2)
mediana = df.median(axis=1).round(2)

print(df)

# Exibir resultados formatados
# Loop que percorre todas as semanas do DataFrame
# enumerate(df.index) obtém o índice númerico(i) e o nome da semana
for i, semana in enumerate(df.index):
    print(f"\n{semana}:")
    print(f" - Média: {media[i]:.2f}")
    print(f" - Mediana: {mediana[i]:.2f}")
    print(f" - Desvio Padrão: {desvio_padrao[i]:.2f}")
```

```python
# Comparação das médias nas semanas

import matplotlib.pyplot as plt

# Criando gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(df.index, media, color=['blue', 'green', 'red', 'purple'])

# Adicionando rótulos
plt.xlabel("Semana")
plt.ylabel("Média de Visitas")
plt.title("Variação das visitas por dia da semana")
plt.ylim(0, max(media) + 10)  # Ajusta o limite do eixo Y
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Exibir gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Eixos x e y
dias_da_semana = [1, 2, 3, 4, 5]
numero_visitas = [10, 20, 15, 25, 30]

# Criando o gráfico de linhas
plt.plot(dias_da_semana, numero_visitas, marker='o', linestyle='-', color='b', label="Visitas")

# Personalizando o gráfico
plt.xlabel("Dias da Semana")
plt.ylabel(" Número de Visitas")
plt.title("Aundiência por dia ")
plt.legend()  # Exibe a legenda
plt.grid(True, linestyle="--", alpha=0.5)  # Adiciona uma grade pontilhada

# Exibir o gráfico
plt.show()
```
